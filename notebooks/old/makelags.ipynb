```python
import pandas
```

```python
d = './TOTALBASE.csv'
data = pandas.read_csv(d)
data[['Code', 'Pub lag']]
```

```python
def make_monthly_lags(x):
    if (x['Code'] == 'TLT') or (x['Code'] == 'IVV') or (x['Code'] == 'TLT_MEAN') or (x['Code'] == 'IVV_MEAN'):
        return int(0)
    else:
        r = str(x['Pub lag']).split(' ')
        if len(r) == 2:
            f, q = r
            if 'day' in q:
                return int(0)
            elif 'month' in q:
                return int(f)
            elif 'quarter' in q:
                return int(int(f) * 3)
            elif 'year' in q:
                return int(int(f) * 12)
            else:
                return pandas.NA
        else:
            return pandas.NA
```

```python
data['s'] = data[['Code', 'Pub lag']].apply(func=make_monthly_lags, axis=1)
data['s']
```

```python
data[pandas.isna(data['s'])]
```

```python
result = data.loc[~pandas.isna(data['s']), ['Code', 's']]
result = result.set_index('Code')
result
```

```python
result.to_dict()
```

```python
import json
with open('lags.json', 'w') as fp:
    json.dump(result.to_dict()['s'], fp)
```
