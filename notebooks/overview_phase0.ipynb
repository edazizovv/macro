```python
import os
import pandas
import seaborn
from matplotlib import pyplot
```

```python
d = "../data/data_folds/"
```

```python
target = "IVV_aggmean_pct"
```

```python
phase = 0
j = 0
postfix = "_RBUSBIS"
```

```python
data_train = pandas.read_excel(f"{d}data_train_ph{phase}{postfix}_{j}.xlsx")
data_train = data_train.rename(columns={data_train.columns[0]: "date"}).set_index("date")
data_train[target] = data_train[target].shift(-1)
data_train
```

```python
data_test = pandas.read_excel(f"{d}data_test_ph{phase}{postfix}_{j}.xlsx")
data_test = data_test.rename(columns={data_test.columns[0]: "date"}).set_index("date")
data_test[target] = data_test[target].shift(-1)
data_test
```

```python
aov = pandas.DataFrame(
    data=
                 {
                     "train": data_train.corr(method=meth)[target], 
                     "test" : data_test.corr(method=meth)[target],
                 },
)
aov = aov[aov.index != target]
aov["z"] = 0
aov = aov.sort_values(by="train")
ax = aov.plot()
```

```python
def find_nth(haystack: str, needle: str, n: int) -> int:
    start = haystack.find(needle)
    while start >= 0 and n > 1:
        start = haystack.find(needle, start+len(needle))
        n -= 1
    return start
```

```python
aov["class"] = aov.index.to_series().apply(func=lambda x: x[find_nth(x, "_", 2)+1:find_nth(x, "_", 3)])
aov["class"] = aov["class"].apply(func=lambda x: x[:x.index("x")-1] if "x" in x else x)
aov["diff"] = (aov["train"] - aov["test"]).abs()
aov["abs"] = aov["train"].abs()
aov = aov.sort_values(by="diff")
aov.head(20)
```

```python
aov_diff = aov[["class", "diff"]].copy().rename(columns={"diff": "value"})
aov_diff["hue"] = "diff"
aov_abs = aov[["class", "abs"]].copy().rename(columns={"abs": "value"})
aov_abs["hue"] = "abs"
aox = pandas.concat(
    [
        aov_diff,
        aov_abs,
    ],
    axis=0,
    ignore_index=True,
)
aox
```

```python
seaborn.boxplot(data=aox, x="class", y="value", hue="hue")
```
